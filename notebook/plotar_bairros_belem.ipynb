# Resumo do Código

O código lê o arquivo shapefile `PA_bairros_CD2022.shp` usando GeoPandas e filtra apenas os bairros pertencentes ao município de Belém (código IBGE `1501402`). Em seguida, cria um mapa interativo com Folium e plota os polígonos dos bairros filtrados, permitindo visualizar os limites municipais diretamente no mapa.


```python
import geopandas as gpd
import os
import folium


```

```python
# -----------------------------------------------------
# OPEN FILE ( from IBGE)
# -----------------------------------------------------
filename="PA_bairros_CD2022.shp"
path='/home/akel/PycharmProjects/Data_ambiental/data/raw/'
arquivo0 = os.path.join(path, filename)

gdf = gpd.read_file(arquivo0)



# Caminho para o arquivo .shp
gdf = gpd.read_file("/home/akel/PycharmProjects/Data_ambiental/data/raw/PA_bairros_CD2022.shp")
gdf.columns
```

```python
#shape file dos bairros de belém
gdf_belem=gdf[gdf['CD_MUN']=='1501402']
gdf_bairros=gdf_belem[['NM_BAIRRO','geometry']] 
```

```python
# salvar o arquivo com shapes
# -----------------------------------------------------
# 4)SALVAR O ARQUIVO
# -----------------------------------------------------
filename="shape_bairros.gpkg"
path='/home/akel/PycharmProjects/Data_ambiental/data/process/'
arquivo = os.path.join(path, filename)

if os.path.exists(arquivo):
    print("GPKG já existe. Não sobrescrever.")
else:
    gdf_bairros.to_file(arquivo, driver="GPKG")
    print("Arquivo GPKG salvo!")
```

```python
# -----------------------------------------------------
# MAPA FOLIUM COM MÚLTIPLOS POLÍGONOS
# -----------------------------------------------------
#OPEN NEW FILE#
filename="shape_bairros.gpkg"
path='/home/akel/PycharmProjects/Data_ambiental/data/process/'
arquivo = os.path.join(path, filename)

gdf_bairros = gpd.read_file(arquivo)


m = folium.Map(location=[-1.45, -48.5], zoom_start=12)
folium.GeoJson(gdf_bairros).add_to(m)
m
```
