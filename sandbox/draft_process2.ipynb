```python
import json
import pandas as pd
import os
```

```python
path="/home/akel/PycharmProjects/Data_ambiental/data/raw/"
filename_a='tabela1699.csv'

dfa = pd.read_csv(path+filename_a,sep=';',skiprows=4)
#removendo o sufixo da variavel bairro  - Belém (PA)
dfa['Bairro'] = dfa['Bairro'].str.replace(' - Belém (PA)', '', regex=False).str.strip()

dfa=dfa.rename(columns={
    "Total": "alfabetização"
})
dfa["alfabetização"] = dfa["alfabetização"].str.replace(",", ".", regex=False).astype(float)
dfa.info()
```

```python
# salvar os novos dados
def salvar_csv(df, caminho):
    if os.path.exists(caminho):
        print(f"⚠️ Arquivo já existe e NÃO será sobrescrito: {caminho}")
    else:
        df.to_csv(caminho, index=False)
        print(f"✅ Arquivo salvo: {caminho}")

path = '/home/akel/PycharmProjects/Data_ambiental/data/process/'
salvar_csv(dfa, path + 'tabela1699_final.csv')
```

```python
path = '/home/akel/PycharmProjects/Data_ambiental/data/process/'
filename_total='tabela_total_final.csv'
df_t = pd.read_csv(path+filename_total)

df_t.info()
```

```python
# # Merge pela chave "Bairro"
df_final = df_t.merge(dfa, on='Bairro', how='left')
df_final
```
