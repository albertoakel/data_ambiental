```python
import folium
import geopandas as gpd

# Carregar o GeoJSON
gdf1 = gpd.read_file("setor_1302.geojson")
gdf2 = gpd.read_file("setor_1304.geojson")
gdf3 = gpd.read_file("setor_1305.geojson")

# Obter centro aproximado do polígono
centro = gdf.geometry.centroid.iloc[0]
mapa = folium.Map(location=[centro.y, centro.x], zoom_start=16)

# Adicionar polígono ao mapa
folium.GeoJson(
    gdf1,
    name="Setor 1302"
).add_to(mapa)
folium.GeoJson(
    gdf2,
    name="Setor 1304"
).add_to(mapa)

folium.GeoJson(
    gdf3,
    name="Setor 1305"
).add_to(mapa)
folium.LayerControl().add_to(mapa)
mapa
```
