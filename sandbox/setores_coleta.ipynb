```python
import geopandas as gpd
from shapely.geometry import Point
import folium

# Lista de setores
lista=['1302','1304','1305','1402','1404','1405','1401','1301','1901','1303','1307','1306']

# Criar lista para armazenar geodataframes
setores = {}

# Carregar todos os setores
for i in lista:
    filename = f"setor_{i}.geojson"
    setores[i] = gpd.read_file(filename)

# Criando o ponto
lon, lat = -48.485, -1.468
pt = Point(lon, lat)

# Testar em qual setor o ponto est√°
setor_encontrado = None

for codigo, gdf in setores.items():
    poligono = gdf.geometry.iloc[0]
    if poligono.contains(pt):
        setor_encontrado = codigo
        break

if setor_encontrado:
    print(f"üìç O ponto est√° dentro do setor **{setor_encontrado}**")
else:
    print("‚ùå O ponto n√£o est√° dentro de nenhum setor")
```

```python
import folium
import geopandas as gpd

# Carregar o GeoJSON
gdf1 = gpd.read_file("setor_1302.geojson")

# Obter centro aproximado do pol√≠gono
centro = gdf1.geometry.centroid.iloc[0]
mapa = folium.Map(location=[centro.y, centro.x], zoom_start=16)


lista=['1302','1304','1305','1402','1404','1405','1401','1301','1901','1303','1307','1306']

# Criar lista para armazenar geodataframes
setores = {}

# Carregar todos os setores
for i in lista:
    filename = f"setor_{i}.geojson"
    setores[i] = gpd.read_file(filename)
    folium.GeoJson(setores[i],name="Setor "+i,style_function=lambda feature: {
            "color": "green",
            "weight": 2,
            "fillColor": "green",
            "fillOpacity": 0.1
        }).add_to(mapa)
folium.LayerControl().add_to(mapa)


# Criando o ponto
lon, lat = -48.495, -1.458
pt = Point(lon, lat)

# Adicionar o ponto ao mapa
folium.Marker(
    location=[lat, lon],     # ordem correta: [lat, lon]
    popup="Ponto testado",
    icon=folium.Icon(color="blue")
).add_to(mapa)

mapa



```
